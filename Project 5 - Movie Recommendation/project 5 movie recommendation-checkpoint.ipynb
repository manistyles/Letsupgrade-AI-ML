# Build Recommender System

```python
c= df['vote_average'].mean()
c
```

```python
q = df['vote_count'].quantile(0.9) #selecting the movies having 90% or above of score
q
```

```python
qm = df[(df['runtime']>150) & (df['revenue'] > 300000000) & (df['homepage'].notna())]
qm
```

```python
qm.shape
```

```python
qm = qm[qm['vote_count'] >= q]
```

```python
qm.shape
```

```python
def wr(x,m=q,c = c):
    v = x['vote_count']
    r = x['vote_average']
    return (v/(v+m)*r)+(m/(m+v)*c) #imdb rating formula
```

```python
qm['score'] = qm.apply(wr,axis = 1) # column wise appending
```

```python
qm.head()
```

```python
qm = qm.sort_values('score',ascending=False)
```
