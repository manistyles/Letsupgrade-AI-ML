```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv("titanic_train.csv")
```

```python
dataset.head(10)
```

```python
from sklearn import preprocessing
from sklearn.naive_bayes import *
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
le = preprocessing.LabelEncoder()
dataset['Sex'] = le.fit_transform(dataset['Sex'])
dataset['Embarked'] = dataset.Embarked.replace('C',1)
dataset['Embarked'] = dataset.Embarked.replace('Q',2)
dataset['Embarked'] = dataset.Embarked.replace('S',3)
```

```python
dataset.columns
```

```python
dataset['Age'].fillna(29,inplace = True)
dataset['Embarked'].fillna(2,inplace = True)
dataset['Embarked'] = dataset['Embarked'].astype(int)
```

```python
dataset.head()
```

```python
def get_accuracy_score(var):
    indi = ['Survived', 'Pclass', 'Sex', 'Age','SibSp','Parch','Fare','Embarked']
    if var in indi:
        indi.remove(var)
    y = dataset[var]
    x = dataset[indi]
    train_x,test_x,train_y,test_y = train_test_split(x,y,test_size = 0.3, random_state = 0)
    glf = GaussianNB()
    y_pred = glf.fit(train_x,train_y).predict(test_x)
    accuracy =accuracy_score(test_y,y_pred,normalize=True)
    confusion = confusion_matrix(test_y,y_pred)
    print("Dependant Variable is",var,"and all remaining variable are independent for this the accuracy score is")
    print(accuracy)
    print()
    print("Dependant Variable is",var,"and all remaining variable are independent for this the confusion matrix is")
    print(confusion)
    print()
    
    
    
dep_var = ['Survived', 'Pclass', 'Sex','SibSp','Parch','Embarked']
for i in dep_var:
    get_accuracy_score(i)
```
