___Project 1: Build Decision Tree(DV-"Survived",IDV-"Age,Gender and Fare") and Prediction___


```python
import pandas as pd
import numpy as np
from sklearn import tree
from sklearn import preprocessing
```

```python
titanic_train = pd.read_csv("titanic_train.csv")
titanic_train.head()
```

```python
titanic_train.isnull().sum()
```

```python
titanic_train['Age'].mean()
```

```python
age = np.where(titanic_train['Age'].isnull(),29,titanic_train['Age'])
titanic_train['Age'] = age
```

```python
titanic_train['Age'].isnull().sum()
```

```python
label_encoder = preprocessing.LabelEncoder()
encoded_sex = label_encoder.fit_transform(titanic_train['Sex'])
```

```python
tree_model = tree.DecisionTreeClassifier(max_depth=4)
```

```python
independent_var = pd.DataFrame([encoded_sex,titanic_train['Age'],titanic_train['Fare']]).T
```

```python
tree_model.fit(X=independent_var,y=titanic_train['Survived'])
```

```python
with open("dtree_titanic.dot",'w') as f:
    f = tree.export_graphviz(tree_model,feature_names=['Sex','Age','Fare'],out_file=f)
```

# Prediction by using Desicion tree

```python
titanic_test = pd.read_csv('titanic_test.csv')
titanic_test.head()
```

```python
titanic_test.isnull().sum()
```

```python
titanic_test['Fare'].fillna(titanic_test['Fare'].mean(),inplace=True)
```

```python
titanic_test['Fare'].isnull().sum()
```

```python
titanic_test.Age.mean()
```

```python
titanic_test['Age'].fillna(30,inplace=True)
```

```python
titanic_test['Age'].isnull().sum()
```

```python
encoded_test_sex = label_encoder.fit_transform(titanic_test['Sex'])
```

```python
test_features = pd.DataFrame([encoded_test_sex,titanic_test['Age'],titanic_test['Fare']]).T
```

```python
test_preds = tree_model.predict(X=test_features)
prediction = pd.DataFrame({'PassengerId':titanic_test['PassengerId'],'Survived':test_preds})
```

```python
prediction.to_csv("Output_model.csv",index=False)
```
