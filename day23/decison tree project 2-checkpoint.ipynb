```python
import numpy as np
import pandas as pd
from sklearn import tree
from sklearn import preprocessing
from sklearn.ensemble import RandomForestClassifier
```

```python
Attrition_data = pd.read_csv("general_data.csv")
Attrition_data.head(10)
```

```python
Attrition_data['TotalWorkingYears'].fillna(11.0,inplace=True)
```

```python
Attrition_data['NumCompaniesWorked'].fillna(2.0,inplace=True)
```

```python
Attrition_data['NumCompaniesWorked'].isna().sum()
```

```python
label_encoder = preprocessing.LabelEncoder()
new_gender = label_encoder.fit_transform(Attrition_data['Gender'])
Attrition_data['Gender'] = new_gender
Attrition_data['Gender']
```

```python
label_encoder = preprocessing.LabelEncoder()
new_Attrition = label_encoder.fit_transform(Attrition_data['Attrition'])
```

```python
features =  ['Age','DistanceFromHome','Education','JobLevel','MonthlyIncome','PercentSalaryHike','TotalWorkingYears','Gender','YearsAtCompany','TrainingTimesLastYear','YearsSinceLastPromotion','YearsWithCurrManager']
```

```python
tree_model = RandomForestClassifier(n_estimators=1000,max_features=2,oob_score=True)
```

```python
tree_model.fit(X=Attrition_data[features],y=new_Attrition)
```

```python
print(tree_model.oob_score_)
```

```python
for feature,imp in zip(features,tree_model.feature_importances_):
    print(feature,imp)
```

___Here Age,DistanceFromHome,MonthlyIncome,TotalWorkingYears are the important features___

# Building a decision tree

```python
model = tree.DecisionTreeClassifier(max_depth=4)
```

```python
IV = pd.DataFrame([Attrition_data['Age'],Attrition_data['DistanceFromHome'],Attrition_data['MonthlyIncome'],Attrition_data['TotalWorkingYears']]).T
```

```python
model.fit(X=IV,y=Attrition_data['Attrition'])
```

```python
with open("P2dtree.dot",'w') as f:
    f = tree.export_graphviz(model,feature_names=['Age','DistanceFromHome','MonthlyIncome','TotalWorkingYears'],out_file=f)
```
