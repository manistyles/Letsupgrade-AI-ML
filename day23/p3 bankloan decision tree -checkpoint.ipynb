```python
import numpy as np
import pandas as pd
from sklearn import tree,preprocessing
from sklearn.ensemble import RandomForestClassifier
```

```python
loan = pd.read_csv("Bank_Personal_Loan_Modelling-1.csv")
```

```python
loan.head()
```

```python
loan.isna().sum()
```

```python
features = ['Age','Experience','Income','Family','CCAvg','Education','Mortgage','Securities Account','CD Account','Online','CreditCard']
```

```python
model = RandomForestClassifier(n_estimators=1000,max_features=2,oob_score=True)
model.fit(X=loan[features],y=loan['Personal Loan'])
```

```python
print(model.oob_score_)
```

```python
for feature,imp in zip(features,model.feature_importances_):
    print(feature,imp)
```

___Here Income,CCAvg,Education are important features to make the decision tree for the given dataset___

# Decision Tree Classification

```python
independent_var = pd.DataFrame([loan['Income'],loan['CCAvg'],loan['Education']]).T
```

```python
tree_model = tree.DecisionTreeClassifier(max_depth=4)
tree_model.fit(X=independent_var,y=loan['Personal Loan'])
```

```python
with open('p3dtreeloan.dot','w') as f:
    f = tree.export_graphviz(tree_model,feature_names=['Income','CCAvg','Education'],out_file=f)
```
